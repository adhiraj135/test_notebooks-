```python
with contextlib.closing(wave.open(path,'r')) as f:
  frames = f.getnframes()
  rate = f.getframerate()
  duration = frames / float(rate)
```

```python

audio = Audio()

def segment_embedding(segment):
  start = segment["start"]
  # Whisper overshoots the end timestamp in the last segment
  end = min(duration, segment["end"])
  clip = Segment(start, end)
  waveform, sample_rate = audio.crop(path, clip)
  return embedding_model(waveform[None])
```

```python
embeddings = np.zeros(shape=(len(segments), 192))
for i, segment in enumerate(segments):
  embeddings[i] = segment_embedding(segment)

embeddings = np.nan_to_num(embeddings)

```

```python

num_speakers=2
```

```python

clustering = AgglomerativeClustering(num_speakers).fit(embeddings)
labels = clustering.labels_
for i in range(len(segments)):
  segments[i]["speaker"] = 'SPEAKER ' + str(labels[i] + 1)
```

```python



def time(secs):
  return datetime.timedelta(seconds=round(secs))

f = open("transcript.txt", "w")

for (i, segment) in enumerate(segments):
  if i == 0 or segments[i - 1]["speaker"] != segment["speaker"]:
    f.write("\n" + segment["speaker"] + ' ' + str(time(segment["start"]))  + '\n')
  f.write(segment["text"][1:] + ' ')
  f.write('\n')
f.close()
```

```python
speaker_diarization=[]
for (i, segment) in enumerate(segments):
  if i == 0 or segments[i - 1]["speaker"] != segment["speaker"]:
    speaker_diarization.append(segment["speaker"] + ' ' + str(time(segment["start"])))
  speaker_diarization.append(segment["text"][1:] + ' ')
```

```python
def speaker_transcript_with_diarization(audio_file):
  # if audio_file[-3:] != 'wav':
  #         subprocess.call(['ffmpeg', '-i', audio_file, 'audio.wav', '-y'])
  #         audio_file = 'audio.wav'
  segments=return_segments(audio_file)
  embeddings = np.zeros(shape=(len(segments), 192))
  for i, segment in enumerate(segments):
      embeddings[i] = segment_embedding(audio_file,segment)

  embeddings = np.nan_to_num(embeddings)
  num_speakers=2

  clustering = AgglomerativeClustering(num_speakers).fit(embeddings)
  labels = clustering.labels_
  for i in range(len(segments)):
      segments[i]["speaker"] = 'SPEAKER ' + str(labels[i] + 1)
  speaker_diarization=[]
  for (i, segment) in enumerate(segments):
    if i == 0 or segments[i - 1]["speaker"] != segment["speaker"]:
        speaker_diarization.append(segment["speaker"] + ' ' + str(time(segment["start"])))
    speaker_diarization.append(segment["text"][1:] + ' ')
  results=''
  for i in speaker_diarization:
     result=translator(i)[0]["translation_text"]
     results=results+result
  return results
```

```python
def hindi_speaker_transcript(audio_file):
    if audio_file[-3:] != 'wav':
          subprocess.call(['ffmpeg', '-i', audio_file, 'audio.wav', '-y'])
          audio_file = 'audio.wav'
    options = {"fp16": False, "language": "hi", "task": "translate"}
    results=model.transcribe(audio_file,**options)
    transcript = results["text"]
    results=translator(transcript)[0]["translation_text"]
    return results
```

```python
def en_speaker_transcript(audio_file):
    if audio_file[-3:] != 'wav':
          subprocess.call(['ffmpeg', '-i', audio_file, 'audio.wav', '-y'])
          audio_file = 'audio.wav'
    options = {"fp16": False, "language": "hi", "task": "translate"}
    results=model.transcribe(audio_file,**options)
    transcript = results["text"]
    return transcript
```
